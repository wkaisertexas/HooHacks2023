# Total Energy Use Since 1950 Consolodation
From the: U.S. Energy Information Administration

```python

renewable = pd.read_csv('raw_data/generation-renewable-sources.csv')
renewable.columns = renewable.iloc[4]
renewable.columns.values[0] = 'year'
renewable = renewable.iloc[5:]
renewable
```

```python
major = pd.read_csv('raw_data/generation-major-source.csv')

major.columns = major.iloc[4]
major.columns.values[0] = 'year'
major = major.iloc[5:]

major
```

```python
total_production = pd.merge(major, renewable, on='year')
total_production.to_json('hosting/public/total_production.json', orient='records')
total_production
```
